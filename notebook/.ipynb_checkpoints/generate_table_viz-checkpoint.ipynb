```python
import pandas as pd
import numpy as np
import json
from jinja2 import Environment, FileSystemLoader
```

```python
def table_data_format(df_orig):
    metadata = []
    df = df_orig
    df.replace(to_replace=["na", "?", np.nan, "missing", "not available",
                           "n/a", "missing value"], value=np.nan, inplace=True)

    for col in df.columns:
        df[col] = pd.to_numeric(df[col], errors='ignore')

    for idx, col in enumerate(df.columns):
        helper = {}
        key = col
        if len(key) > 15:
            key = col[:15]

        helper['ascend'] = 0
        helper['name'] = key
        if(df.dtypes[col] == "object"):
            df[col].replace(to_replace=["na", "?", np.nan, "missing", "not available",
                                        "n/a", "missing value"], value="missing value", inplace=True)
            helper['datatype'] = 'string'
            helper['values'] = df[col].unique().tolist()
        elif(df.dtypes[col] == "int64"):
            helper['datatype'] = 'int'
            helper['min_val'] = df[col].min()
            helper['max_val'] = df[col].max()
        elif(df.dtypes[col] == "float64"):
            helper['datatype'] = 'float'
            helper['min_val'] = df[col].min()
            helper['max_val'] = df[col].max()
        metadata.append(helper)

    df.replace(to_replace=["na", "?", np.nan, "missing", "not available",
                           "n/a", "missing value"], value="missing value", inplace=True)
    data_array = []
    for row in range(df.shape[0]):
        aux = []
        for col in range(df.shape[1]):
            aux.append(df.iloc[row, col])
        data_array.append(aux)
    return metadata, data_array
```

```python
path = r'C:\Users\Luis_Dutra\3D Objects\datasets\enem\enem_2019_for_table_viz.csv'

# Ler quantidade determinada de linhas do dataset
df = pd.read_csv(path)

df
```

```python
def corrigir_palavras(nome):
    nome = nome.replace('.', '').replace('ç', 'c').replace('ô', 'o').replace('é', 'e').replace('ã','a').replace('ê','e').replace('í','i').replace('ì','').replace('Ì','I').replace('Í','I').replace('í','i').replace('ì','').replace('ó','o').replace('ò','o').replace('ú','u').replace('u','ù').replace('á','a')
            
    return nome
```

```python
df['TP_SEXO'] = df['TP_SEXO'].apply(corrigir_palavras)
df['TP_ESTADO_CIVIL'] = df['TP_ESTADO_CIVIL'].apply(corrigir_palavras)
df['TP_COR_RACA'] = df['TP_COR_RACA'].apply(corrigir_palavras)
df['TP_NACIONALIDADE'] = df['TP_NACIONALIDADE'].apply(corrigir_palavras)
df['SG_UF_NASCIMENTO'] = df['SG_UF_NASCIMENTO'].apply(corrigir_palavras)
df['TP_ST_CONCLUSAO'] = df['TP_ST_CONCLUSAO'].apply(corrigir_palavras)
df['TP_ANO_CONCLUIU'] = df['TP_ANO_CONCLUIU'].apply(corrigir_palavras)
df['TP_ESCOLA'] = df['TP_ESCOLA'].apply(corrigir_palavras)
df['TP_ENSINO'] = df['TP_ENSINO'].apply(corrigir_palavras)
df['IN_TREINEIRO'] = df['IN_TREINEIRO'].apply(corrigir_palavras)
df['TP_DEPENDENCIA_ADM_ESC'] = df['TP_DEPENDENCIA_ADM_ESC'].apply(corrigir_palavras)
df['TP_LOCALIZACAO_ESC'] = df['TP_LOCALIZACAO_ESC'].apply(corrigir_palavras)
df['TP_SIT_FUNC_ESC'] = df['TP_SIT_FUNC_ESC'].apply(corrigir_palavras)
df['IN_SURDEZ'] = df['IN_SURDEZ'].apply(corrigir_palavras)
df['IN_SURDO_CEGUEIRA'] = df['IN_SURDO_CEGUEIRA'].apply(corrigir_palavras)
df['IN_DEFICIENCIA_FISICA'] = df['IN_DEFICIENCIA_FISICA'].apply(corrigir_palavras)
df['IN_DEFICIENCIA_MENTAL'] = df['IN_DEFICIENCIA_MENTAL'].apply(corrigir_palavras)
df['IN_DEFICIT_ATENCAO'] = df['IN_DEFICIT_ATENCAO'].apply(corrigir_palavras)
df['IN_DISLEXIA'] = df['IN_DISLEXIA'].apply(corrigir_palavras)
df['IN_DISCALCULIA'] = df['IN_DISCALCULIA'].apply(corrigir_palavras)
df['IN_AUTISMO'] = df['IN_AUTISMO'].apply(corrigir_palavras)
df['IN_GESTANTE'] = df['IN_GESTANTE'].apply(corrigir_palavras)
df['TP_LINGUA'] = df['TP_LINGUA'].apply(corrigir_palavras)
df['TP_STATUS_REDACAO'] = df['TP_STATUS_REDACAO'].apply(corrigir_palavras)
df['Q001'] = df['Q001'].apply(corrigir_palavras)
df['Q002'] = df['Q002'].apply(corrigir_palavras)
df['Q006'] = df['Q006'].apply(corrigir_palavras)
df['Q025'] = df['Q025'].apply(corrigir_palavras)
```

```python
# Adiciona o  metadata_table_viz e dataarray_table_viz no html

file_loader = FileSystemLoader(r'C:\pandas-profiling\pandas_profiling\templates')
env = Environment(loader=file_loader)

template = env.get_template("table.html")

keys, data_array = table_data_format(df)

output = template.render(metadata_table_viz=keys, dataarray_table_viz=data_array)

# C:\pandas-profiling\notebook
with open("test.html", "w") as result_file:
    result_file.write(output)
```
